```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
def generate_data():
    x = np.arange(-60, 60, 2)
    y = 8*x**2 + 6*x**3 + np.random.randn(60)*1000
    return x, y
df, df1 = generate_data()
df = df/max(df)
df1 = df1/max(df1)
#plot the data
fig, ax = plt.subplots()
ax.scatter(df, df1)
```

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression 
```

```python
n = 6
df_expanded = np.expand_dims(df, axis=1)
poly = PolynomialFeatures(n)
# returns: [1, x, x^2, x^3]
df_expanded = poly.fit_transform(df_expanded)
```

```python
# Initializing and training the Linear Regression model
model_lr = LinearRegression()
model_lr.fit(df.reshape(-1, 1), df1.reshape(-1, 1))

# Predicting the values from the model
y_pred_lr = model_lr.predict(df.reshape(-1, 1))

# Plotting the predictions
plt.figure(figsize=(10, 5))
plt.scatter(df, df1, s = 15)
plt.plot(df, y_pred_lr, color='r', label='Linear Regression')
plt.xlabel('Predictor')
plt.ylabel('Target')
plt.show()
```

```python
inp = Input((n+1)) 
#since one of the features is 1, we need an extra input
out = Dense(1)(inp)
model = Model(inputs=inp, outputs=out)
model.compile(optimizer=Adam(lr=1e-3), loss="mean_squared_error")
```

```python
model.fit(df_expanded, df1, epochs=100)
plt.scatter(df, df1)
plt.plot(df, model.predict(df_expanded), color="red")
```
